```python
%%sql
CREATE TABLE IF NOT EXISTS Players (
    player_id INTEGER PRIMARY KEY,
    name      TEXT,
    team      TEXT
);
```

```python
%%sql
CREATE TABLE IF NOT EXISTS Stats (
    player_id  INTEGER,
    yards      INTEGER,
    touchdowns INTEGER
);
```

```python
%%sql
-- player_id 5 (Tee Higgins) intentionally has NO stats row
-- player_id 6 in Stats intentionally has NO matching player
INSERT INTO Players VALUES
    (1, 'Ja''Marr Chase', 'Bengals'),
    (2, 'Joe Burrow',     'Bengals'),
    (3, 'Jerry Jeudy',    'Browns'),
    (4, 'Deshaun Watson', 'Browns'),
    (5, 'Tee Higgins',    'Bengals');

INSERT INTO Stats VALUES
    (1, 1200, 10),
    (2,  850,  8),
    (3,  950,  6),
    (4,  720,  5),
    (6,  300,  2);
```

```python
%%sql
SELECT p.name, s.yards, s.touchdowns
FROM Players AS p
JOIN Stats AS s
  ON p.player_id = s.player_id;
```

```python
%%sql
SELECT p.name, s.yards, s.touchdowns
FROM Players AS p
LEFT JOIN Stats AS s
  ON p.player_id = s.player_id;

```

```python
%%sql
SELECT name
FROM Players
WHERE player_id IN (
  SELECT player_id FROM Stats
);
```

```python
%%sql
SELECT name
FROM Players
WHERE player_id NOT IN (
  SELECT player_id FROM Stats
);
```

```python
%%sql
SELECT a.name AS player1, b.name AS player2, a.team
FROM Players AS a
JOIN Players AS b
  ON a.team = b.team
  AND a.player_id < b.player_id;
```

```python
%%sql

SELECT name FROM Players WHERE team = 'Bengals'
UNION
SELECT p.name
FROM Players AS p
JOIN Stats AS s ON p.player_id = s.player_id
WHERE s.yards > 800;
```
