```python
%%sql
DROP TABLE IF EXISTS users;

CREATE TABLE users (
    id INTEGER PRIMARY KEY,
    username TEXT,
    password TEXT
);

INSERT INTO users (username, password) VALUES
('admin', 'supersecret'),
('student', 'learning123'),
('guest', 'welcome');
```

```python
import IPython
def login(username, password):
    query = f"SELECT * FROM users WHERE username = '{username}' AND password = '{password}';"
    print("Running query:")
    print(query)
    result = IPython.get_ipython().run_line_magic('sql', query)
    return result
```

```python
import sqlite3

def login(username, password):
    conn = sqlite3.connect("security_lab.db")
    cursor = conn.cursor()

    query = f"SELECT * FROM users WHERE username = '{username}' AND password = '{password}';"

    print("Running query:")
    print(query)

    cursor.execute(query)
    results = cursor.fetchall()

    conn.close()

    print("Results:")
    return results
```

```python
import sqlite3

def secure_login(username, password):
    conn = sqlite3.connect("security_lab.db")
    cursor = conn.cursor()

    query = "SELECT * FROM users WHERE username = ? AND password = ?;"
    print("Running secure query (parameterized)")

    cursor.execute(query, (username, password))
    results = cursor.fetchall()

    conn.close()
    return results
```

```python
secure_login('admin', "' OR '1'='1")
```
