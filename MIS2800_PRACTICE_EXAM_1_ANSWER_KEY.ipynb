***Online Level Transaction Processing (OLTP) is the "front-line" system. It is optimized for speed and accuracy in handling day-to-day business operations—like a customer buying a jersey or an athlete registering for a race. It focuses on CRUD (Create, Read, Update, Delete) in the moment.***

# 3. In a star schema, the center table is called:

A. Dimension table
B. Lookup table
C. **Fact table**
D. Bridge table



***The center table of a star schema is the Fact Table. It acts as the central hub, storing quantitative, numerical, and measurable data (measures or metrics) such as sales amounts, quantities, or profit. This table is surrounded by dimension tables and typically contains foreign keys that link to them.***

**Analogy: *Think of the Fact Table as the box score of a game. It contains the measurable, quantitative data (the "facts" or numbers, like points scored or items sold). The surrounding Dimension Tables (the points of the star) contain the descriptive data like names, dates, and locations.***

# 4. First Normal Form (1NF) requires:

A. No foreign keys
B. **No repeating groups**
C. No primary keys
D. Only one table



***1NF is about "Atomicity," meaning data cannot be broken down any further. If a cell contains a list (like "QB, Captain"), it violates 1NF. Every cell must contain exactly one piece of information—one box, one jersey.***

Just as an atom was historically considered the smallest unit of matter that couldn't be split without losing its identity, an atomic value in a database is the smallest unit of data that should be stored in a single cell.

In science, if you split an atom, you’re no longer dealing with the same element. In SQL, if you "split" a truly atomic value (like the number 7), you just have meaningless fragments.

A cell containing "QB, Captain" is like a molecule—it's a compound made of two distinct parts. 1NF requires us to break that molecule down into its individual atoms.

"Think of 1NF as the 'Atomic Phase.' Just as an atom is the most singular form of matter, an atomic value is the most singular form of data.

If a cell contains a list (like 'QB, Captain'), it’s a molecule, not an atom. To reach 1NF, we must break it down until every cell holds exactly one 'atom' of information.

Note: What is 'atomic' depends on your goal. A full name 'Joe Burrow' might be an atom in a simple list, but in a professional database, the atoms are actually 'Joe' and 'Burrow.'"
